```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

l = 3
n = 4
m = 512
decoder = 'onms'

mc_df = pd.read_csv("FromMatlabScript/h_"+str(l)+"_"+str(n)+"_"+str(m)+".sprs.mc." +
                    decoder +".results", skipinitialspace=True)
ffh_df = pd.read_csv("FromMatlabScript/h_"+str(l)+"_"+str(n)+"_"+str(m)+".sprs.ffh." + 
                     decoder + ".results", skipinitialspace=True)

snr = mc_df['SNR']
mc = mc_df['FER']
snr2 = ffh_df['SNR']
ffh = ffh_df['FER']
plt.figure(figsize=(10,8))
plt.plot(snr, mc, '-o', snr2, ffh, '-D');
plt.title("H with params l={}, n={}, m={} ".format(l,n,m))
plt.legend(["MC", "FFH"])
plt.xlabel("SNR")
plt.grid(True)
plt.ylabel("FER")
plt.yscale('log')
plt.savefig("h_"+str(l)+"_"+str(n)+"_"+str(m)+"_" + decoder + ".png");
```
